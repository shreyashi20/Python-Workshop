# Merge and join 

```python
df1=pd.DataFrame({'employee':['Bob','Jake','Lias','Sue'],
                'group':['Accounting','Engineering','Engineering','HR']})
df2=pd.DataFrame({'employee':['Bob','Jake','Lias','Sue'],
                  'hire_data':[2004,2005,2006,2014]})
df1
```

```python
df2
```

```python
df3=pd.merge(df1,df2)
df3
```

```python
df4=pd.DataFrame({'group':['Accounting','Engineering','HR'],'Supervisor':['Carly','Guidi','Steve']})
df4
```

```python
display(df1,df2,df3,df4)
```

```python
pd.merge(df3,df4)  
# or we can use display directly
```

```python
display(pd.merge(df3,df4))
```

```python
df5=pd.DataFrame({'group':['Accounting','Accounting','Engineering','Engineering','HR','HR'],
                  'skills':['math','spreadsheet','coding','linux','spreadsheet','organisation']})
df5
```

```python
display(pd.merge(df1,df5))
```

```python
df6=pd.DataFrame({'name':['Bob','jake','Lisa','Sue'],
                  'salary':['7000','90000','28444','66647']})
df6
```

```python
display(df1,df6,pd.merge(df1,df6,left_on="employee", right_on="name"))
```

```python
pd.merge(df1,df6,left_on="employee", right_on="name").drop('name',axis=1)
```

```python
df7=pd.DataFrame({'name':['Hitesh', 'Taral','Pratik'],
                 'food':['fish','egg','meat']},columns=['name','food'])
df8=pd.DataFrame({'name':['Pratik','Pukhraj'],
                 'drink':['wine','beer']},columns=['name','drink'])
display(df7,df8,pd.merge(df7,df8))

```

```python
pd.merge(df7,df8,how='outer')
```

```python
pd.merge(df7,df8,how='left')
```

```python
pd.merge(df7,df8,how='right')
```

```python
df9=pd.DataFrame({'name':['Rose','Judy','Max','Bing'],
                 'rank': [1,2,3,4]})
df10=pd.DataFrame({'name':['Rose','Judy','Max','Bing'],
                 'rank': [1,2,3,4]})
display(df9,df10,pd.merge(df9,df10,on="name"))
```

```python
display(df9,df10,pd.merge(df9,df10,on="name", suffixes=["_L","_R"]))
```
